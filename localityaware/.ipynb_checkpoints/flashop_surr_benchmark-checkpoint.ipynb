```python
import numpy as np
import pickle
import os

# Define the file path for saving/loading the data
data_file = "flash_simulation_data_sim.pkl"

# Define parameter ranges
flash_1_range = np.linspace(6, 12, 5)
flash_2_range = np.linspace(0.5, 4, 5)

# Check if the data file already exists
if os.path.exists(data_file):
    print("Loading existing data...")
    with open(data_file, "rb") as f:
        data = pickle.load(f)
    assSim = Refrig2DrumConstraintHeatExConstDummy(AspenFile="../FlashOperation/FlashOperation_HeatExchanger.bkp",
                                               wdpath="../FlashOperation",
                                               visibility=False,
                                               Penalty=1e3)
else:
    print("Generating new data...")
    assSim = Refrig2DrumConstraintHeatExConstDummy(AspenFile="../FlashOperation/FlashOperation_HeatExchanger.bkp",
                                               wdpath="../FlashOperation",
                                               visibility=False,
                                               Penalty=1e3)

    data = []
    for flash_1 in flash_1_range:
        for flash_2 in flash_2_range:
            x_unflat = assSim.unflatten_params([float(flash_1), float(flash_2)])
            data.append([flash_1, flash_2, assSim.run_obj(x_unflat)])

    # Save the data for future use
    with open(data_file, "wb") as f:
        pickle.dump(data, f)
    print(f"Data saved to {data_file}")

# Data is now available
print("Data loaded, number of entries:", len(data))
data = np.array(data)
```

```python
# **Initialize Model & Datasets**
old_dataset = OldDataSet(data, k=5)
new_dataset = NewDataSet(k=5)
model = MLP(2, [5, 5, 5], 1)

# **Run Optimization**
x_path, y_path = optimize_surrogate_model(model, 
                                          old_dataset,
                                          new_dataset, 
                                          assSim, 
                                          optim_steps=100, 
                                          N_s=2, 
                                          lr={"model learning rate": 1e-3, "input learning rate": 1e-3}, 
                                          merge_interval=20,
                                          x_init = torch.tensor([1.0, 1.0]),
                                            epoch = 3,
                                            batch_size_old = 256,
                                            batch_size_new = 2
                                         )
```

```python
plt.plot(range(100), y_path, label='Objective Value')
plt.xlabel('Iteration')
plt.ylabel('Objective Value')
plt.title('Objective Value over Iterations')
plt.legend()
plt.show()
```

```python
# Save the data to a file
with open("./xpathypath_sim.pkl", "wb") as file:
    pickle.dump([x_path,y_path], file)
```

```python
#open from download folder
with open("../FlashOperation/flash_simulation_data.pkl", "rb") as file:
    flashsimdata = pickle.load(file)
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Convert flashsimdata to numpy array for easier manipulation
flashsimdata = np.array(flashsimdata)

# Extract the columns for Flash 1, Flash 2, and the objective values
flash_1 = flashsimdata[:, 0]
flash_2 = flashsimdata[:, 1]
objective_values = flashsimdata[:, 2]

# Create a grid for Flash 1 and Flash 2
flash_1_grid, flash_2_grid = np.meshgrid(np.unique(flash_1), np.unique(flash_2))

# Reshape the objective values to match the grid shape
objective_values_grid = objective_values.reshape(flash_1_grid.shape)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(flash_1_grid, flash_2_grid, objective_values_grid, cmap='viridis')

# Add labels and title
ax.set_xlabel('Flash 1')
ax.set_ylabel('Flash 2')
ax.set_zlabel('Objective Value')
ax.set_title('Response Surface of Flash Simulation Data')

# Show the plot
plt.show()
```

```python
# Plot x path on top of the surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(flash_1_grid, flash_2_grid, objective_values_grid, cmap='viridis', alpha=0.7)

# Plot the x_path
x_path = np.array(x_path)
ax.plot(x_path[:, 0], x_path[:, 1], y_path, 'r-o', label='Optimization Path')

# Add labels and title
ax.set_xlabel('Flash 1')
ax.set_ylabel('Flash 2')
ax.set_zlabel('Objective Value')
ax.set_title('Optimization Path on Response Surface')
ax.legend()

# Show the plot
plt.show()
```
