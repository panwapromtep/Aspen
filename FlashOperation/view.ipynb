# View Pickle Files
This notebook allows you to load and view the data stored in the pickle files generated by `test_genetic.py` and `rosenbrock_ga.py`.

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

# Define file paths for the pickle files
test_genetic_obj_values_pickle_file = "obj_values_vs_generations.pkl"
test_genetic_population_pickle_file = "population_distribution.pkl"
rosenbrock_obj_values_pickle_file = "rosenbrock_obj_values_vs_generations.pkl"
rosenbrock_population_pickle_file = "rosenbrock_population_distribution.pkl"

# Load objective values (average, minimum) and Aspen calls vs. generations for test_genetic
with open(test_genetic_obj_values_pickle_file, "rb") as f:
    test_genetic_obj_values_data = pickle.load(f)
    print("Test Genetic Objective Values and Aspen Calls vs. Generations:")
    print(test_genetic_obj_values_data)

# Load population distribution over iterations for test_genetic
with open(test_genetic_population_pickle_file, "rb") as f:
    test_genetic_population_data = pickle.load(f)
    print("\nTest Genetic Population Distribution Over Iterations:")
    for entry in test_genetic_population_data:
        print(f"Generation {entry['Generation']}: {entry['Population']}")

# Load objective values (average, minimum) and Aspen calls vs. generations for rosenbrock_ga
with open(rosenbrock_obj_values_pickle_file, "rb") as f:
    rosenbrock_obj_values_data = pickle.load(f)
    print("\nRosenbrock Objective Values and Aspen Calls vs. Generations:")
    print(rosenbrock_obj_values_data)

# Load population distribution over iterations for rosenbrock_ga
with open(rosenbrock_population_pickle_file, "rb") as f:
    rosenbrock_population_data = pickle.load(f)
    print("\nRosenbrock Population Distribution Over Iterations:")
    for entry in rosenbrock_population_data:
        print(f"Generation {entry['Generation']}: {entry['Population']}")

# Plot average and minimum objective values vs. generations for rosenbrock_ga
plt.figure(figsize=(8, 6))
plt.plot(rosenbrock_obj_values_data['Generations'], rosenbrock_obj_values_data['Avg_Obj_Values'], marker='o', label='Average Objective Value', color='blue')
plt.plot(rosenbrock_obj_values_data['Generations'], rosenbrock_obj_values_data['Min_Obj_Values'], marker='s', label='Minimum Objective Value', color='green')
plt.xlabel('Generation')
plt.ylabel('Objective Value')
plt.title('Rosenbrock Objective Value vs. Generation')
plt.legend()
plt.grid(True)
plt.show()

# Plot Aspen calls vs objective values for rosenbrock_ga
plt.figure(figsize=(8, 6))
plt.plot(rosenbrock_obj_values_data['Aspen_Calls'], rosenbrock_obj_values_data['Avg_Obj_Values'], marker='o', label='Average Objective Value', color='blue')
plt.plot(rosenbrock_obj_values_data['Aspen_Calls'], rosenbrock_obj_values_data['Min_Obj_Values'], marker='s', label='Minimum Objective Value', color='green')
plt.xlabel('Number of Aspen Calls')
plt.ylabel('Objective Value')
plt.title('Rosenbrock Objective Value vs. Aspen Calls')
plt.legend()
plt.grid(True)
plt.show()

# Visualize population distribution over iterations for rosenbrock_ga
plt.figure(figsize=(8, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(rosenbrock_population_data)))
for idx, entry in enumerate(rosenbrock_population_data):
    generation = entry['Generation']
    population = entry['Population']
    x1 = [individual[0] for individual in population]
    x2 = [individual[1] for individual in population]
    plt.scatter(x1, x2, color=colors[idx], label=f'Generation {generation}')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Rosenbrock Population Distribution Over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```

## Rosenbrock Problem that shows distribution over iterations

```python
iteration = 1
# Visualize population distribution over iterations for rosenbrock_ga
colors = plt.cm.viridis(np.linspace(0, 1, len(rosenbrock_population_data)))
for idx, entry in enumerate(rosenbrock_population_data):
    generation = entry['Generation']
    population = entry['Population']
    x1 = [individual[0] for individual in population]
    x2 = [individual[1] for individual in population]
    plt.figure(figsize=(8, 6))
    plt.scatter(x1, x2, color=colors[idx], label=f'Generation {generation}')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.xlim([-15, 15])
    plt.ylim([-15, 15])
    plt.title(f'Gen {iteration} Rosenbrock Population Distribution Over Iterations')
    plt.legend()
    plt.grid(True)
    plt.show()
    iteration += 1
```
