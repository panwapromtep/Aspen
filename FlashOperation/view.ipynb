# View Pickle Files
This notebook allows you to load and view the data stored in the pickle files generated by `test_genetic.py`.

```python
import pickle
import matplotlib.pyplot as plt

# Define file paths for the pickle files
obj_values_pickle_file = "obj_values_vs_generations.pkl"
population_pickle_file = "population_distribution.pkl"

# Load objective values (average, minimum) and Aspen calls vs. generations
with open(obj_values_pickle_file, "rb") as f:
    obj_values_data = pickle.load(f)
    print("Objective Values and Aspen Calls vs. Generations:")
    print(obj_values_data)

# Load population distribution over iterations
with open(population_pickle_file, "rb") as f:
    population_data = pickle.load(f)
    print("\nPopulation Distribution Over Iterations:")
    for entry in population_data:
        print(f"Generation {entry['Generation']}: {entry['Population']}")

# Plot average and minimum objective values vs. generations
plt.figure(figsize=(10, 6))
plt.plot(obj_values_data['Generations'], obj_values_data['Avg_Obj_Values'], marker='o', label='Average Objective Value', color='blue')
plt.plot(obj_values_data['Generations'], obj_values_data['Min_Obj_Values'], marker='s', label='Minimum Objective Value', color='green')
plt.xlabel('Generation')
plt.ylabel('Objective Value')
plt.title('Objective Value vs. Generation')
plt.legend()
plt.grid(True)
plt.show()

# Plot Aspen calls vs minimum objective values
plt.figure(figsize=(10, 6))
plt.plot(obj_values_data['Aspen_Calls'], obj_values_data['Avg_Obj_Values'], marker='o', label='Average Objective Value', color='blue')
plt.plot(obj_values_data['Aspen_Calls'], obj_values_data['Min_Obj_Values'], marker='o', label='Minimum Objective Value', color='red')
plt.xlabel('Number of Aspen Calls')
plt.ylabel('Objective Value')
plt.title(' Objective Value vs. Aspen Calls')
plt.legend()
plt.grid(True)
plt.show()

# Visualize population distribution over iterations
plt.figure(figsize=(10, 6))
for entry in population_data:
    generation = entry['Generation']
    population = entry['Population']
    x1 = [individual[0] for individual in population]
    x2 = [individual[1] for individual in population]
    plt.scatter(x1, x2, label=f'Generation {generation}')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Population Distribution Over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
